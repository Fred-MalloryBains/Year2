```python
import numpy as np

import math
import matplotlib.pyplot as plt

a = [] #create a list of size 0
a.append(1/math.sqrt(3)) # append 1/root(3), the first term of the sequence
function(a,a[len(a)-1]) # called the function to calculate PI values, parameters are the list and its' size
```

```python
def function (a, num):
    while (len(a) < 50):
        new = ((math.sqrt(num **2 + 1 )-1)/num)
        a.append(new)
        function(a,a[len(a)-1])
    return a
```

### Results

### Interpretation

[*If we look at the algorithm, the problem is that it depends on the divisor of the previous operation, however we can algerbraiacly rearrange the forumla to get it so the divisor never reaches 0, yet they are equicalent functions *]

### Improved implementation


```python
def new_function (a,num):
    while (len(a) < 27):
        new = (num / (math.sqrt(num**2 + 1) + 1))
        a.append(new)
        new_function(a,a[len(a)-1])
    return a
```

### Improved results

[*The second algorithm is much better to calculate up larger values of n as there is no zero division and therefore no error.*]

### Interpretation
